```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)              #initialize the pseudorandom number generator
                                #with a specified seed.

# Number of samples to take
sample_size = 40

# Number of experiments to conduct
num_experiments = 1000

# Create subplots for each distribution
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Central Limit Theorem Demonstration', fontsize=19)

# Different distributions to test
distributions = [
    np.random.uniform(0, 1, sample_size),
    np.random.exponential(1, sample_size),
    np.random.normal(0, 1, sample_size),
    np.random.poisson(3, sample_size)
]
                                                  #SURAJ JADHAV Q2
for i, distribution in enumerate(distributions):
    means = []

    # Conduct 1000 experiments
    for _ in range(num_experiments):
        sample = np.random.choice(distribution, sample_size)
        sample_mean = np.mean(sample)
        means.append(sample_mean)

    # Plot the distribution of means
    ax = axs[i // 2, i % 2]
    ax.hist(means, bins=30, density=True, alpha=0.7, color='red')
    ax.set_title(f'Distribution {i+1}')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
